```python
import numpy as np 
import pandas as pd 
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits, on="title")
```

What to remain

```python
#genres
#id
#keywords
#title
#overview
#production_companies
#cast
#crew
#release dates

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew','release_date']]
```

```python
movies.head()
```

Data Preprocessing

```python
#inoreder to merge all except movie_id and title
#we need to preprocess the data inside each column to a mergable and understandable data and delete missing data 
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies = movies.reset_index(drop=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
    names = []
    for i in ast.literal_eval(obj):
        names.append(i['name'])
    return names   
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies = movies[movies['genres'].apply(lambda x : len(x)>0)]
movies = movies.reset_index(drop=True)
```

```python
movies['genres']
```

```python
movies.head()
```

```python
movies['keywords'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['keywords']
```

```python
movies.head()
```

```python
def convert_cast(obj):
    names = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 5:
            names.append(i['character'])
            counter += 1
        else:
            break    
    return names
```

```python
movies['cast'].apply(convert_cast)
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['release_date'] = pd.to_datetime(movies['release_date']).dt.year
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) # convert overview into a list of words
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags','genres','release_date']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

Text Vectorization

```python
# A score is to be calcualted to check the similarity between the texts
# Like no of same words
# But this is not that simple or efficient so vectorization of text is used 
```

```python
# In text vectorization, text is converted into vertors using the method bag of words another method is there as well tfidf but here we use the first one
# all tags for a movie will be concatenated and most used words are chosen out of them 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# we can observe that we get some similar words 
# so we combine all the similar words to make a single word
# this is done using a library called "nltk" which is a natural language processing libarary
# it gives root word for every word like loving,loved -> love
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# function to convert text into their stem forms
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# do vectorization again with updated tags list
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
# now we have vectors for each movie with each vector is of length 5000 which is most used 5000 words
# now we need to find the distances of vectors from each other
# but vectors as dimensions go higher euclidean distances fail often 
# therefore we find the cosinal distance between the vectors i.e the angles between the vectors
# dist is inversely proportional to similarity
# there is a function available in sci-kit learn which is cosine similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(similarity[0],reverse=True)[1:11]
# sorting from similarity of a movie in descending order such that we get first 10 similar movies other than the original one
# but here comes the problem that is we lsoe the index no of the movies when we sort
# to solve it we use enumerate 
```

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x:x[1])
# sorting enumerated list with respect to column 1 not column 0
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie ].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:20] # we need first 5 other than the given one
    
    for i in movies_list :
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Interstellar')
```
